```python
import psycopg2
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Create an engine instance
alchemyEngine = create_engine('postgresql+psycopg2://myuser:pass@0.0.0.0:5432/mydb')
# Connect to PostgreSQL server
dbConnection = alchemyEngine.connect()
```

```python
# Read data from PostgreSQL database table and load into a DataFrame instance
USDT_BTC = pd.read_sql("select * from \"USDT_BTC\"", dbConnection, index_col  = 'id')
USDT_ZEC = pd.read_sql("select * from \"USDT_ZEC\"", dbConnection)
USDT_ETH = pd.read_sql("select * from \"USDT_ETH\"", dbConnection)
USDT_XMR = pd.read_sql("select * from \"USDT_XMR\"", dbConnection)
USDT_ETC = pd.read_sql("select * from \"USDT_ETC\"", dbConnection)
USDT_DASH = pd.read_sql("select * from \"USDT_DASH\"", dbConnection)
USDT_LTC = pd.read_sql("select * from \"USDT_LTC\"", dbConnection)
```

```python
ls1 = {'USDT_BTC':USDT_BTC ,'USDT_ZEC':USDT_ZEC ,'USDT_ETH':USDT_ETH ,'USDT_XMR':USDT_XMR ,'USDT_ETC':USDT_ETC ,'USDT_DASH':USDT_DASH ,'USDT_LTC':USDT_LTC}
```

```python
fig, ax  = plt.subplots(9, sharey='row' ,figsize=(15 ,20))

for exp_id in ls1.keys():
    ax[0].plot(ls1[exp_id]['last'], label=exp_id)
ax[0].legend()
ax[0].set(title='last');

for exp_id in ls1.keys():
    ax[1].plot(ls1[exp_id]['lowestAsk'], label=exp_id)
ax[1].legend()
ax[1].set(title='lowestAsk');

for exp_id in ls1.keys():
    ax[2].plot(ls1[exp_id]['highestBid'], label=exp_id)
ax[2].legend()
ax[2].set(title='highestBid');

for exp_id in ls1.keys():
    ax[3].plot(ls1[exp_id]['percentChange'], label=exp_id)
ax[3].legend()
ax[3].set(title='percentChange');

for exp_id in ls1.keys():
    ax[4].plot(ls1[exp_id]['baseVolume'], label=exp_id)
ax[4].legend()
ax[4].set(title='baseVolume');

for exp_id in ls1.keys():
    ax[5].plot(ls1[exp_id]['quoteVolume'], label=exp_id)
ax[5].legend()
ax[5].set(title='quoteVolume');

for exp_id in ls1.keys():
    ax[6].plot(ls1[exp_id]['isFrozen'], label=exp_id)
ax[6].legend()
ax[6].set(title='isFrozen');

for exp_id in ls1.keys():
    ax[7].plot(ls1[exp_id]['high24hr'], label=exp_id)
ax[7].legend()
ax[7].set(title='high24hr');

for exp_id in ls1.keys():
    ax[8].plot(ls1[exp_id]['low24hr'], label=exp_id)
ax[8].legend()
ax[8].set(title='low24hr');
```
